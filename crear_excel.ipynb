```python
wb = load_workbook('ingredientes2016.xlsx')

# añadimos
pestaña = wb['Reporte'] # con esto identificamos la hoja


min_col = wb.active.min_column
max_col = wb.active.max_column
min_row = wb.active.min_row
max_row = wb.active.max_row

print(min_col, max_col, min_row, max_row)

```

```python
# vamos a añadir la media al final
# para ello tendremos que llamar a cada celda por columna(letra) + fila(numero)
# ord('A') = 65, ord('Z') = 90
i = 65 # el valor que le sumamos para poner la 'A'
letra_delante = ""
letra = 66
c = 0 # indica la columna


while c < 32:
    # la fila donde lo vamos a poner es la 59
    columna = letra_delante+f"{chr(letra)}"
    # Calculamos el promedio de cada fila
    pestaña[f"{columna}59"] = f'=SUM({columna}5:{columna}58)/53'
    #pestaña[f"{columna}59"].style = 5
    c += 1
    letra += 1
    if letra == 91:
        # cuando nos pasamos de la Z añadimos la A delante, hasta que nos volvamos a pasar and so on
        letra_delante = chr(i)
        i += 1
        letra = 65
pestaña['A59'] = 'Media' # le ponemos nombre a la fila

wb.save('ingredientes2016.xlsx')
```

## Graficos

```python
barchart = BarChart()

# le decimos que la data esta en esa pestaña, y le indicamos de donde a donde está
data = Reference(pestaña, min_col=min_col + 1, max_col=max_col, min_row=min_row, max_row=max_row)
# solo cogemos las categorias (la columna es la misma)
categorias = Reference(pestaña, min_col=min_col, max_col=min_col, min_row=min_row + 1, max_row=max_row)

barchart.add_data(data, titles_from_data=True)

barchart.set_categories(categorias)

pestaña.add_chart(barchart, 'AI34') # le decimos que añada el barchart a la pestaña, y le decimos donde ponerlo
barchart.title = 'Pizzas pedidas' # titulo
barchart.style = 5
```

```python
# vamos a hacer u n grafico para la pizza mas pedida: thai_ckn = AE
barchart = BarChart()

barchart.type = "col"


data = Reference(pestaña, min_col=max_col-2, max_col=max_col-2, min_row=min_row, max_row=max_row)

categorias = Reference(pestaña, min_col=min_col, max_col=min_col, min_row=min_row + 1, max_row=max_row)


barchart.add_data(data, titles_from_data=True)

barchart.y_axis.title = 'Numero pizzas pedidas'
barchart.x_axis.title = 'Semanas'

pestaña.add_chart(barchart, 'AI1') # le decimos que añada el barchart a la pestaña, y le decimos donde ponerlo
barchart.title = 'Pizzas thai_ckn pedidas a lo largo del año' # titulo
barchart.style = 10

#wb.save('ingredientes2016.xlsx')
```

## Hacemos un diagrama de sectores


```python
projected_pie = ProjectedPieChart()
projected_pie.type = "pie"
projected_pie.splitType = "val" # split by value
labels = Reference(pestaña, min_col=2, max_col=31, min_row=4, max_row=4)
data = Reference(pestaña, min_col=2, max_col=31, min_row=59, max_row=59)
projected_pie.add_data(data, titles_from_data=True)
projected_pie.set_categories(labels)

pestaña.add_chart(projected_pie, "AI1")

from copy import deepcopy
projected_bar = deepcopy(projected_pie)
projected_bar.type = "bar"
projected_bar.splitType = 'pos' # split by position

pestaña.add_chart(projected_bar, "AI20")

wb.save('ingredientes2016.xlsx')
```
