```python
wb = load_workbook('ingredientes2016.xlsx')
pestaña = wb['Reporte'] # con esto identificamos la hoja

min_col = wb.active.min_column
max_col = wb.active.max_column
min_row = wb.active.min_row
max_row = wb.active.max_row

print(min_col, max_col, min_row, max_row)
```

# Añadir grafico

```python
barchart = BarChart()

# le decimos que la data esta en esa pestaña, y le indicamos de donde a donde está
data = Reference(pestaña, min_col=min_col + 1, max_col=max_col, min_row=min_row, max_row=max_row)
# solo cogemos las categorias (la columna es la misma)
categorias = Reference(pestaña, min_col=min_col, max_col=min_col, min_row=min_row + 1, max_row=max_row)

barchart.add_data(data, titles_from_data=True)

barchart.set_categories(categorias)

pestaña.add_chart(barchart, 'AI34') # le decimos que añada el barchart a la pestaña, y le decimos donde ponerlo
barchart.title = 'Pizzas pedidas' # titulo
barchart.style = 5

wb.save('ingredientes2016.xlsx')
```
